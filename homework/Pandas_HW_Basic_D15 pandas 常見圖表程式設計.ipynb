```python
"""
題目 : 將資料夾中 boston.csv 讀進來，並用圖表分析欄位。

畫出箱型圖，並判斷哪個欄位的中位數在 300~400 之間？
畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？
"""
#Part 01 讀 data
import numpy as np
import pandas as pd

#STEP 01 先把要讀取的檔案放到一樣階層的資料夾裡面
#利用 usecols 指定要讀取的欄位
boston_data = pd.read_csv ('boston.csv')
print('boston_data: \n', boston_data)
```

```python
import numpy as np
import pandas as pd

#畫出箱型圖，並判斷哪個欄位的中位數在 300~400 之間？
#箱型圖
print('BOX: \n', boston_data.boxplot(figsize = (12,6))) #p.s. figsize 可以調整 figure size!!!

#中位數
print('median: \n', boston_data.median())
print('betwwen 300 to 400', boston_data.median().between(300, 400))

#畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？
```

```python
#畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？
boston_data.plot.scatter (x='NOX', y='DIS')
print('應該是反比吧?')
```
