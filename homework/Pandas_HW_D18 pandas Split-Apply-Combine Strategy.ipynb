```python
"""
題目 : 運用下列分數資料分析

找出全年級各科成績最高分與最低分?
找出數學班平均最高的班級?
分析全校女生與男生國文平均差幾分?
"""
import numpy as np
import pandas as pd
score_df = pd.DataFrame([[1,50,80,70,'boy',1],[2,60,45,50,'boy',2],[3,98,43,55,'boy',1],[4,70,69,89,'boy',2],
                         [5,56,79,60,'girl',1],[6,60,68,55,'girl',2],[7,45,70,77,'girl',1],[8,55,77,76,'girl',2],
                         [9,25,57,60,'girl',1],[10,88,40,43,'girl',3],[11,25,60,45,'boy',3],[12,80,60,23,'boy',3],
                         [13,20,90,66,'girl',3],[14,50,50,50,'girl',3],[15,89,67,77,'girl',3]],
                         columns=['student_id','math_score','english_score','chinese_score','sex','class'])

print(score_df)

#找出全年級各科成績最高分與最低分?
print('math_max:', score_df['math_score'].max())
print('math_min:', score_df['math_score'].min())
print('english_max:', score_df['english_score'].max())
print('english_min:', score_df['english_score'].min())
print('chinese_max:', score_df['chinese_score'].max())
print('chinese_min:', score_df['chinese_score'].min())

#找出數學班平均最高的班級?
#重新把 student_id 定義成 index 的 data
score_df_new = score_df.set_index('student_id')
print('new data: \n', score_df_new, '\n\n')
q2=score_df_new.groupby(['class']).agg(('mean','max'))
print('q2: \n', q2, '\n\n')

#分析全校女生與男生國文平均差幾分?
#q3=score_df_new.groupby(['sex', 'class']).agg(('mean','max'))
q3=score_df_new.groupby(['sex']).agg(('mean','max'))
print('q3: \n', q3, '\n \n')

#pick up: .loc
print(q3.loc['boy','chinese_score']-q3.loc['girl','chinese_score'])
```
