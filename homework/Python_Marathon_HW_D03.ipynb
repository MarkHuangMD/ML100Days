```python
# 1. [簡答題] 請比較 np.zeros 和 np.empty 產生出來的陣列有何差異？為什麼要設計兩種方法？

a = np.zeros((4,6), dtype='float64')
b = np.empty((3,5), dtype='float64')
print('a:', a)
print('b:', b)

"""
如果想要產稱一個由 0 所組成的矩陣，可以使用 np.zeors(...) 的方法。除此之外，在文件中有另外一個類似的方法是 np.empty(...)：
呼叫 zeros()、ones() 函式，可以依照傳入的形狀引數，建立元素全為 0、全為 1 的陣列。
使用 empty() 函式則是不需要給定起始值，但是可以建立給定形狀的陣列，元素值則會隨機給定。
"""
c=np.empty((2,3), dtype='float64')
print('c:', c)

"""
#empty會建立很接近0的數值, zeros會建立0的數值
#empty必須讓使用者輸入數值，zeros是建立可以使用的0在array

np.zeros((2, 3)) 會產生真正的 0；np.empty((2, 3)) 會產生一個近似於 0 的陣列。
np.empty 可以用於只是想要初始化，但之後就會數值覆寫過的形況。
"""
```
