```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 
tips = sns.load_dataset("tips")
#fmri = sns.load_dataset("fmri")

print('tips = ', tips.head(5), '\n \n')
#print('fmri = ', fmri.head(5), '\n \n')

"""
Q1 , hue  sex
x="total_bill", y="tip", hue="sex", style="time"
"""

# x, y, hue : names of variables in data or vector data, optional
# In seaborn, the hue parameter determines which column in the data frame should be used "for colour encoding"

q1 = sns.lmplot(x="total_bill", y="tip", data=tips)
print (q1)
# dding `hue="smoker" tells seaborn you want to colour the data points for smoker and non-smoker differently.
q1_hue = sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
print(q1_hue)

# 
#  style !!
sns.set(style="darkgrid")
sns.relplot(x="total_bill", y="tip", hue="sex", style="time", data=tips)
```

```python
# 
fmri = sns.load_dataset("fmri")
print('fmri = \n', fmri.head(5), '\n \n')

"""
Q2 fmri, 
Q2  x="timepoint", y="signal", hue="event", style="region"
"""
# data: fmri
sns.set(style="darkgrid")
q2 = sns.relplot (x="timepoint", y="signal", hue="event", style="region", data = fmri)
print(q2)
# relplot() - relationship plot
```

```python
# fmri,  sns.regplot
# x="timepoint", y="signal"
# regression plot
q2_reg = sns.regplot(x="timepoint", y="signal", data=fmri)
print(q2_reg)
```
